Menampilkan: daftar sheet, kolom tiap sheet, dan 10 baris contoh data.

```python
import pandas as pd
from pathlib import Path

# Path file: jalan dari sirama-checker atau dari root proyek
path = Path("030226 MK SIRAMA V2.xlsx")
if not path.exists():
    path = Path("sirama-checker/030226 MK SIRAMA V2.xlsx")

xl = pd.ExcelFile(path)

print("Jumlah sheet:", len(xl.sheet_names))
print("\nNama sheet:")
for i, name in enumerate(xl.sheet_names, 1):
    print(f"  {i}. {name}")
```

```python
# Per sheet: kolom + 10 contoh baris
for sheet_name in xl.sheet_names:
    df = pd.read_excel(xl, sheet_name=sheet_name)
    print("=" * 60)
    print(f"SHEET: {sheet_name}")
    print("=" * 60)
    print(f"Kolom ({len(df.columns)}): {list(df.columns)}")
    print(f"\n10 contoh data (baris pertama):\n")
    display(df.head(10))
    print()
```
